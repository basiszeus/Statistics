## Chi Square Distribution

### PDF

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parameters
df = 2  # degrees of freedom
x = np.linspace(0, 20, 100)
pdf = chi2.pdf(x, df)
```

```python
df = 10  
lifespan = 10  #
pdf_value = chi2.pdf(lifespan, df) # 0.0877
print(f'The probability density function value for a lifespan of {lifespan} days with {df} degrees of freedom is: {pdf_value:.4f}')

x = np.linspace(0, 30, 100)
pdf = chi2.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Chi-Squared Distribution PDF (df={df})', color='blue')
plt.axvline(lifespan, color='red', linestyle='--', label=f'Lifespan = {lifespan} days')
plt.title('Chi-Squared Distribution PDF')
plt.xlabel('Lifespan (days)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```

---

### CDF

```python
cdf = chi2.cdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, cdf, label=f'Chi-Squared Distribution CDF (df={df})', color='orange')
plt.title('Chi-Squared Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid()
plt.show()
```

```python
df = 8  # degrees of freedom
x = 12  # lifespan value

cdf_value = chi2.cdf(x, df)
cdf_value 
```

```python
x_values = np.linspace(0, 20, 500)
cdf_values = chi2.cdf(x_values, df)

plt.figure(figsize=(10, 6))
plt.plot(x_values, cdf_values, color='b', label=f'Chi-Squared CDF with {df} df')
plt.axvline(x=x, color='r', linestyle='--', label=f'Lifespan = {x}')
plt.axhline(y=cdf_value, color='g', linestyle='--', label=f'CDF at lifespan {x} = {cdf_value:.4f}')
plt.title('CDF of Chi-Squared Distribution')
plt.xlabel('Lifespan (days)')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid(True)
plt.show()
```
