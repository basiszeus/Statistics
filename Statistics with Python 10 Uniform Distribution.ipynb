# Continuous uniform distribution

```python
continuous_uniform_distribution = uniform(loc=a, scale=b)
```

## PDF

```python
continuous_uniform_pdf = continuous_uniform_distribution.pdf(x)
print(continuous_uniform_pdf)
```

```python
#2 
pdf = 1/b-a
pdf
```

```python
#visual

plt.plot(x, continuous_uniform_pdf)
plt.xlabel('X')
plt.ylabel('Probability')
plt.show()
```

## CDF

```python
continuous_uniform_cdf = continuous_uniform_distribution.cdf(x)
print(continuous_uniform_cdf)
```

```python
#visual
plt.plot(x, continuous_uniform_cdf)
plt.xlabel('X')
plt.ylabel('Cumulative Probability')
plt.show()
```

# Discrete uniform distribution

```python
a=1
b=6
x = np.arange(a, b+1) #1 to 6
print(x)
```

```python
discrete_uniform_distribution = randint(a, b+1)
```

##  PMF

```python
discrete_uniform_pmf = discrete_uniform_distribution.pmf(x)
print(discrete_uniform_pmf)
plt.plot(x, discrete_uniform_pmf, 'bo', ms=8)
plt.vlines(x, 0, discrete_uniform_pmf, colors='b', lw=5, alpha=0.5)
plt.xlabel('X')
plt.ylabel('Probability')
plt.show()
```

## CDF

```python
discrete_uniform_cdf = discrete_uniform_distribution.cdf(x)
print(discrete_uniform_cdf)
plt.plot(x, discrete_uniform_cdf, 'bo', ms=8)
plt.xlabel('X')
plt.ylabel('Cumulative Probability')
plt.show()
```
