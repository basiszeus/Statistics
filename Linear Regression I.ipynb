Finding the Slop and the Intercept with Scipy
</h1><center/>

```python
from scipy import stats

# Given data
salary = [100000, 120000, 80000, 115000, 70000, 55000, 60000, 65000, 75000]
years_of_experience = [6, 5, 4, 6, 3, 1, 1, 2, 2]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(years_of_experience, salary)

# Print regression coefficients
print("Slope (b1):", slope) # Slope (b1): 10885.416666666668
print("Intercept (b0):", intercept) # Intercept (b0): 45937.49999999999
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 6))
plt.scatter(years_of_experience, salary, color='blue', label='Data Points')
plt.plot(years_of_experience, intercept + slope * np.array(years_of_experience), color='red', label='Regression Line')
plt.title('Scatter Plot of Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.xlim(0, 10) 
plt.ylim(0, 150000)
plt.show()
```

Finding the Slop and the Intercept using the formula
</h1><center/>

## Finding the Slop

```python
from scipy.stats import pearsonr

# Data
salary = [100000, 120000, 80000, 115000, 70000, 55000, 60000, 65000, 75000]
years_of_experience = [6, 5, 4, 6, 3, 1, 1, 2, 2]

# Calculate correlation coefficient
correlation, _ = pearsonr(years_of_experience, salary)

# Calculate the std
from scipy.stats import tstd

std_dev_x = tstd(years_of_experience)
std_dev_y = tstd(salary)

# Calculate slope using the formula
slope = correlation * (std_dev_y / std_dev_x)

# Print slope (b1)
print("Slope (b1):", slope)
```

## Calculate the Intercept

```python
import numpy as np

# Given data
salary = [100000, 120000, 80000, 115000, 70000, 55000, 60000, 65000, 75000]
years_of_experience = [6, 5, 4, 6, 3, 1, 1, 2, 2]

# Calculate intercept using the formula
intercept = np.mean(salary) - slope * np.mean(years_of_experience)

# Print intercept (b0)
print("Intercept (b0):", intercept)
```
