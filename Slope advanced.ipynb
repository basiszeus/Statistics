```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

### Slope of a curve

```python
# Define the symbol and the function
x = sp.symbols('x')
f = x**2

# Calculate the derivative
f_prime = sp.diff(f, x)

# Point at which to find the slope
x_point = 2
slope = f_prime.subs(x, x_point)

# Print the derivative and the slope at x = 2
print(f"Derivative of f(x): {f_prime}")
print(f"Slope at x = {x_point}: {slope}")
```

```python
# Plot the function and the tangent line
x_vals = np.linspace(0, 4, 400)
y_vals = [f.subs(x, val) for val in x_vals]

plt.plot(x_vals, y_vals, label='$f(x) = x^2$')
plt.plot(x_point, f.subs(x, x_point), 'ro', label='Point of Tangency')

# Calculate the tangent line
y_tangent = [slope * (val - x_point) + f.subs(x, x_point) for val in x_vals]
plt.plot(x_vals, y_tangent, '--', label='Tangent at $x=2$')

plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.title('Slope of the Curve at $x=2$')
plt.grid(True)
plt.show()
```

###  Gradient

```python
# Define the function
x, y = sp.symbols('x y')
f = x**2 + y**2

# Calculate the gradient
grad_f = [sp.diff(f, var) for var in (x, y)]
print(f"Gradient: {grad_f}")

# Convert the gradient to a numpy function
grad_f_func = sp.lambdify((x, y), grad_f)

# Create a meshgrid for plotting
x_vals = np.linspace(-5, 5, 20)
y_vals = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x_vals, y_vals)

# Evaluate the gradient at each point
U, V = grad_f_func(X, Y)

# Plot the function surface
fig, ax = plt.subplots()
contours = ax.contour(X, Y, X**2 + Y**2, levels=20, cmap='viridis')
ax.quiver(X, Y, U, V, color='red')

# Customize and show the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Gradient Field')
plt.show()
```
