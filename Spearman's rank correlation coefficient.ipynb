## No Tie

```python
import numpy as np
from scipy.stats import spearmanr
import seaborn as sns

# Sample data with no ties
x_no_ties = [10, 20, 30, 40, 50]
y_no_ties = [5, 15, 25, 35, 45]

# Compute Spearman's rank correlation for no ties
correlation_no_ties, p_value_no_ties = spearmanr(x_no_ties, y_no_ties)

print(f"Spearman's rank correlation coefficient (no ties): {correlation_no_ties}")
print(f"P-value (no ties): {p_value_no_ties}")

sns.scatterplot(x = x_no_ties, y = y_no_ties)
```

## Tie

```python
# Sample data with ties
x_with_ties = [10, 20, 30, 20, 10]
y_with_ties = [20, 30, 40, 30, 10]

# Compute Spearman's rank correlation for data with ties
correlation_with_ties, p_value_with_ties = spearmanr(x_with_ties, y_with_ties)

print(f"Spearman's rank correlation coefficient (with ties): {correlation_with_ties}")
print(f"P-value (with ties): {p_value_with_ties}")

sns.scatterplot(x = x_with_ties, y = y_with_ties)
```
