```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

### Rate of visits per minute

```python
lambda_per_minute = 0.05 #3/60
```

### Q1

```python
# Create an exponential distribution with the scale parameter as 1/lambda
exponential_distribution = expon(scale=1/lambda_per_minute)

# Define the time range for the plot
time_range = np.linspace(0, 60, 500)  # 0 to 60 minutes

# Calculate the PDF for the time range
pdf_values = exponential_distribution.pdf(time_range)

# Calculate the CDF for the time range
cdf_values = exponential_distribution.cdf(time_range)

# Q1: Probability that a visitor will arrive in less than 10 minutes
prob_less_than_10 = exponential_distribution.cdf(10)
prob_less_than_10_per = round(prob_less_than_10 * 100)
print (prob_less_than_10_per,'%')

# Visualization for Q1
plt.figure(figsize=(10, 6))
plt.plot(time_range, pdf_values, label='PDF of Exponential Distribution')
plt.fill_between(time_range, 0, pdf_values, where=(time_range <= 10), alpha=0.5, color='skyblue', label='P(T < 10 minutes)')
plt.title('Q1: Probability that a visitor will arrive in less than 10 minutes')
plt.xlabel('Time (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

### Q2

```python
# Q2: Probability that the next visitor will arrive after 20 minutes
prob_more_than_20 = 1 - exponential_distribution.cdf(20)
prob_more_than_20_per = round(prob_more_than_20 * 100)
print (prob_more_than_20_per,'%') #37 %

# Visualization for Q2
plt.figure(figsize=(10, 6))
plt.plot(time_range, pdf_values, label='PDF of Exponential Distribution')
plt.fill_between(time_range, 0, pdf_values, where=(time_range >= 20), alpha=0.5, color='lightcoral', label='P(T > 20 minutes)')
plt.title('Q2: Probability that the next visitor will arrive after 20 minutes')
plt.xlabel('Time (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

### Q3

```python
prob_exactly_10 = 0

# Visualization for Q3
plt.figure(figsize=(10, 6))
plt.plot(time_range, pdf_values, label='PDF of Exponential Distribution')
plt.axvline(10, color='orange', linestyle='--', label='T = 10 minutes (Exact probability is 0)')
plt.title('Q3: Probability that the next visitor will arrive in exactly 10 minutes')
plt.xlabel('Time (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

### Q4

```python
# Q4: Probability of having a new customer between 5 and 10 minutes
prob_between_5_and_10 = exponential_distribution.cdf(10) - exponential_distribution.cdf(5)
prob_between_5_and_10_per = round(prob_between_5_and_10 * 100)
print (prob_between_5_and_10_per,'%') #17 %


# Visualization for Q4
plt.figure(figsize=(10, 6))
plt.plot(time_range, pdf_values, label='PDF of Exponential Distribution')
plt.fill_between(time_range, 0, pdf_values, where=((time_range >= 5) & (time_range <= 10)), alpha=0.5, color='lightgreen', label='P(5 < T < 10 minutes)')
plt.title('Q4: Probability of having a new customer between 5 and 10 minutes')
plt.xlabel('Time (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
