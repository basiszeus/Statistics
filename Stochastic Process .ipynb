```python
import numpy as np
import matplotlib.pyplot as plt

num_days = 100

initial_price = 100

daily_returns = np.random.normal(0, 1, num_days)

stock_prices = np.zeros(num_days)
stock_prices[0] = initial_price

for t in range(1, num_days):
    stock_prices[t] = stock_prices[t-1] + daily_returns[t]

plt.figure(figsize=(10, 6))
plt.plot(stock_prices, label='Simulated Stock Price')
plt.title('Simulated Stock Price Over Time (Random Walk)')
plt.xlabel('Day')
plt.ylabel('Price')
plt.legend()
plt.show()
```
