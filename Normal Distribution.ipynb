```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### Drawing Normal Distribution

```python
# Parameters for the normal distribution
mean = 0       # Mean of the distribution
std_dev = 1    # Standard deviation of the distribution
size = 1000    # Number of data points

# Generate random data from a normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=size)
data
```

```python
# Create a range of values for the x-axis
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the PDF for these x values
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Plot the histogram of the generated data
plt.figure(figsize=(12, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Histogram')

# Plot the PDF
plt.plot(x, pdf, 'k', linewidth=2, label='Normal Distribution PDF')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution: Histogram and PDF')
plt.legend()

# Show plot
plt.show()
```

### PDF

```python
# Parameters for the normal distribution
mu = 170  # Mean height in cm
sigma = 10  # Standard deviation in cm

# Generate values for x-axis (heights)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate PDF valu
pdf_values = norm.pdf(x, mu, sigma)

# Plotting the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_values, label=f'Normal Distribution\nMean = {mu}, SD = {sigma}')
plt.fill_between(x, pdf_values, alpha=0.2)
plt.title('Probability Density Function (PDF) of Normal Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate PDF value at x
pdf_values = norm.pdf(x, mu, sigma)

specific_value = 180
pdf_at_specific_value = norm.pdf(specific_value, mu, sigma)

# Convert PDF value to percentage
pdf_at_specific_value_percent = pdf_at_specific_value * 100

print(f'The PDF value at x = {specific_value} is {pdf_at_specific_value_percent}%')
```

```python
# Plotting the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_values, label='Normal Distribution PDF', color='blue')
plt.axvline(specific_value, color='red', linestyle='--', label=f'PDF at {specific_value} = {pdf_at_specific_value_percent:.2f}%')

# Annotate the PDF value on the graph
plt.text(specific_value, pdf_at_specific_value + 0.0005, f'{pdf_at_specific_value_percent:.2f}%', color='red')

# Titles and labels
plt.title('Normal Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Density (%)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

### CDF

```python
# Calculate the CDF at 180 cm
x_value = 180
cdf_value = norm.cdf(x_value, mu, sigma)
cdf_at_specific_value_percent = cdf_value * 100

# Calculate PDF values
pdf_values = norm.pdf(x, mu, sigma)

# Plotting the PDF and highlighting the area under the curve for CDF up to 180 cm
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_values, label=f'Normal Distribution\nMean = {mu}, SD = {sigma}')
plt.fill_between(x, pdf_values, where=(x <= x_value), alpha=0.2, color='blue', label=f'CDF up to {x_value} cm = {cdf_value:.2f}')
plt.axvline(x=x_value, color='red', linestyle='--', label=f'x = {x_value} cm')

plt.title('PDF of Normal Distribution with CDF Highlighted')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

print(f'The CDF value at x = {x_value} is {cdf_at_specific_value_percent}%')
```

### Finding Z Score

```python
import scipy.stats as stats

# Define the Z-score
z_score = -0.55

# Calculate the area to the left of the Z-score
area_left = stats.norm.cdf(z_score)

# Calculate the area to the right of the Z-score
area_right = 1 - area_left

# Print the results
print(f'Area to the left of Z-score {z_score}: {area_left:.5f}')
print(f'Area to the right of Z-score {z_score}: {area_right:.5f}')
```

### Example

```python
mu = 4
sigma = 0.5
observations = [2.5, 3, 3.5, 4, 4.5, 5, 5.5]

z_scores = [(x - mu) / sigma for x in observations]

z_scores

# Calculate the proportion of having a flight time > 3.5 hours
z_greater_than_3_5 = (3.5 - mu) / sigma
proportion_greater_than_3_5 = 1 - stats.norm.cdf(z_greater_than_3_5)

# Proportion of having a flight time between 3 and 3.5 hours
z_3 = (3 - mu) / sigma
z_3_5 = (3.5 - mu) / sigma

proportion_between_3_and_3_5 = stats.norm.cdf(z_3_5) - stats.norm.cdf(z_3)

z_scores, proportion_greater_than_3_5, proportion_between_3_and_3_5
```
