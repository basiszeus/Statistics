```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

# PDF

```python
#We need x , mu and sigma for normal distribution
def normal_dis_pdf(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp((-1 / 2)* ((x- mu / sigma) ** 2))
```

```python
#Let's test it
# x will be a np.array, you can also do it with a signle value
# Mu would be 0 and std is equal to 1

x = np.array([1,2,3])
```

```python
normal_dis_pdf(x,0,1)
```

```python
# You can also jsut do 
norm.pdf(x,loc=0,scale=1)
```

```python
norm.pdf(x)
#By default loc is equal to 0 and scale is equal to 1
```

```python
#Let's do somevisualization
#We will have values that start fro -7 to 7 with 10000 values

x = np.linspace(-7,7,10000)
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(x,normal_dis_pdf(x,0,1))
plt.show()
```

# CDF

```python
norm.cdf(x,loc=0,scale=1)
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(x,norm.cdf(x,0,1))
plt.show()
```
