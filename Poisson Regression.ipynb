```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

## Sample dataset: daily ER visits with some features

```python
df = pd.DataFrame({
    "visits": [40, 55, 28, 50, 60, 35, 45],
    "flu_season": [1, 1, 0, 1, 1, 0, 0],   # 1 if flu season, 0 otherwise
    "heat_wave": [0, 0, 1, 0, 0, 1, 0],    # 1 if heat wave, 0 otherwise
    "weekend": [0, 0, 1, 0, 0, 1, 0]       # 1 if weekend, 0 if weekday
})
df
```

## Add constant term for intercept

```python
X = sm.add_constant(df[["flu_season", "heat_wave", "weekend"]])
y = df["visits"]
```

## Fit Poisson regression

```python
model = sm.GLM(y, X, family=sm.families.Poisson()).fit()
```

## Print model summary

```python
print(model.summary())
```

## Predict new values

```python
new_day = pd.DataFrame({
    "const": [1],
    "flu_season": [1],
    "heat_wave": [0],
    "weekend": [1]
})   
```

```python
predicted_visits = model.predict(new_day)
print(f"Predicted ER visits: {predicted_visits[0]:.1f}")
```
