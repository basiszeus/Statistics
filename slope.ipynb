```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Define the points
x1, y1 = 0, 200
x2, y2 = 3, 500

# Calculate the slope
slope = (y2 - y1) / (x2 - x1)

# Create the figure and axis
fig, ax = plt.subplots()

# Plot the points
ax.plot([x1, x2], [y1, y2], marker='o')

# Annotate the points
ax.text(x1, y1, f'({x1}, {y1})', fontsize=12, ha='right')
ax.text(x2, y2, f'({x2}, {y2})', fontsize=12, ha='left')

# Draw the line
ax.plot([x1, x2], [y1, y2], 'b-')

# Add a grid
ax.grid(True)

# Add title and labels
ax.set_title('Slope Example')
ax.set_xlabel('Distance (km)')
ax.set_ylabel('Elevation (m)')

# Display the slope
plt.figtext(0.5, 0.01, f'Slope = {slope} meters per kilometer', ha='center', fontsize=6)

# Show the plot
plt.show()
```

###  Example II

```python
data = {
    'Marketing Spend ($)': [10000, 15000, 20000, 25000, 30000],
    'Sales Revenue ($)': [50000, 65000, 80000, 95000, 110000]
}
df = pd.DataFrame(data)
df
```

```python
# Extract x and y values
x = data['Marketing Spend ($)']
y = data['Sales Revenue ($)']

# Calculate the slope
x1, y1 = x[0], y[0]
x2, y2 = x[-1], y[-1]
slope = (y2 - y1) / (x2 - x1)

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the points
ax.plot(x, y, marker='o', linestyle='-', color='b')

# Annotate the points with larger font
for i, (x_val, y_val) in enumerate(zip(x, y)):
    ax.text(x_val, y_val, f'({x_val}, {y_val})', fontsize=14, ha='right')

# Add a grid
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Add title and labels with larger font
ax.set_title('Marketing Spend vs. Sales Revenue', fontsize=18)
ax.set_xlabel('Marketing Spend ($)', fontsize=16)
ax.set_ylabel('Sales Revenue ($)', fontsize=16)

# Display the slope with larger font
plt.figtext(0.5, 0.02, f'Slope = {slope:.2f} dollars of sales per dollar of marketing spend', ha='center', fontsize=14)

# Show the plot
plt.show()
```
