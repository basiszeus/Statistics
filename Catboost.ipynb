```python
import pandas as pd

data = {
    "Row": [1, 2, 3, 4, 5],
    "City": ["Draria", "Algiers", "Draria", "Oran", "Draria"],
    "Target": [1, 0, 1, 0, 1],  
}

df = pd.DataFrame(data)

alpha = 0.5

sum_of_past_targets = 2 
max_past_count = 2  

encoded_value = (sum_of_past_targets + alpha) / (max_past_count + 1)

print("Encoded Value for 'Draria' in Row 5:", encoded_value)
```

```python
!pip install catboost
import pandas as pd
from catboost import CatBoostRegressor, Pool
```

```python
data = {
    'Residence': ['Cyanite', 'Azurite', 'Agate', 'Celestine', 'Larimar', 'Azurite', 'Cyanite'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'West', 'South'],
    'Engagement': [120, 95, 150, 80, 110, 100, 90]
}
df = pd.DataFrame(data)

df
```

```python
cat_features = ['Residence', 'Region']
target = 'Engagement'
train_data = Pool(df[cat_features], label=df[target], cat_features=cat_features)
model = CatBoostRegressor(iterations=500, learning_rate=0.1, depth=6, verbose=100)
model.fit(train_data)
```

```python
preds = model.predict(df[cat_features])
preds
```

```python
preds = model.predict(df[cat_features])

df['Predicted_Engagement'] = preds

df
```

## Classification

```python
data_classification = {
    'Residence': ['Cyanite', 'Azurite', 'Agate', 'Celestine', 'Larimar', 'Azurite', 'Cyanite'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'West', 'South'],
    'Engagement_Level': ['High', 'Low', 'High', 'Low', 'High', 'Low', 'Low']
}
df_class = pd.DataFrame(data_classification)
df_class
```

```python
from catboost import CatBoostClassifier

cat_features = ['Residence', 'Region']
target = 'Engagement_Level'


train_data = Pool(df_class[cat_features], label=df_class[target], cat_features=cat_features)
model_class = CatBoostClassifier(iterations=500, learning_rate=0.1, depth=6, verbose=100)
model_class.fit(train_data)
```

```python
preds_class = model_class.predict(df_class[cat_features])

df_class['Predicted_Engagement_Level'] = preds_class

df_class
```

```python
probs_class = model_class.predict_proba(df_class[cat_features])
probs_class
```
